```python
# -*- coding: UTF-8 -*-
import re
import numpy as np
import requests as rq
import os
import pandas as pd
import newspaper
from collections import Counter
import warnings
warnings.filterwarnings('ignore')

def get_article(url, lang):
    """Get text of article from url and language"""
    a = newspaper.build_article(url, lang=lang)
    a.download()
    a.parse()
    return a

lang_dict = {
    "movies-danish - movies-danish.tsv": "da",
    "movies-dutch - movies-dutch.tsv": "nl",
    "movies-english - movies-english.tsv": "en",
    "movies-finnish - movies-finnish.tsv": "fi",
    "movies-norwegian - movies-norwegian.tsv": "no",
    "movies-swedish - movies-swedish.tsv": "sv",
    "music-danish - music-danish.tsv": "da",
    "music-german - music-german.tsv": "de",
    "music-norwegian - music-norwegian.tsv": "no", 
    "music-polish - music-polish.tsv": "en",
    "music-swedish - music-swedish.tsv": "sv"
}

bad_domains = Counter()
ok_domains = Counter()

for f in os.listdir("../data/quotes"):

    if f == ".DS_Store":
        continue

    print "processing:", f
    print "... language", lang_dict[f]

    if f in os.listdir("../data/reviews"):
        print "... already processed"
        continue

    dat = pd.read_csv(
        "../data/quotes/%s" % f,
        names=["url", "quote", "score", "orig_score"],
        delimiter="\t"
    )

    reviews = []
    for item in dat.iterrows():

        domain = re.findall(r".*\://(?:www.)?([^\/]+)", item[1]['url'])[0]

        print ".",

        if bad_domains[domain] * 2 > ok_domains[domain] and bad_domains[domain] > 5:
            print "?"
            reviews.append("")
            continue

        a = get_article(item[1]['url'], lang_dict[f])
        cf_quote = "".join(re.findall(r"\w+", item[1]['quote'][3:-3]))  # 'compact
        cf_review = "".join(re.findall(r"\w+", a.text))                 # form'

        if cf_quote not in cf_review:
            print "Page broken", domain
            bad_domains.update([domain])

            reviews.append("")
            continue

        print item[0] * 1.0 / dat.shape[0]

        ok_domains.update([domain])
        reviews.append(a.text)

    dat.loc[:, 'review'] = pd.Series(reviews, index=dat.index)
    dat.to_csv("../data/reviews/%s" % f, encoding="utf-8")

    print "Successfully wrote", f, "to review_quotes"
```
